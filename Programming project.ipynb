#### 3. Outliers

Now we will check our outliers; there are varios methods in order to check outliers. Firstly, we will use visualization for general overview. We will create new dataframe, which only consistence of numeric values, for this reason we will use Numpy library.

```python
df_numeric = df_cleaned.select_dtypes(include=np.number)
df_numeric
```

```python
df_numeric_1 = df_numeric.drop("All Fuels", axis=1)
```

```python
plt.figure(figsize=(18, 10))
rows = 5  
cols = 4 
for i, col in enumerate(df_numeric_1, 1):
    plt.subplot(rows, cols, i) 
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

This boxplot provides us a general overview of the data distribution of data, and highlight potential outliers across categories. As we see the boxplots crowded with outliers, making it hard to interpret the data effectively. This can happen when there are large variations in the data or outliers that are far from the rest of the values..Now, we will look more details of our outliers. For this purpose we will calculate IQR.

```python
def detect_outliers_iqr(df_numeric, local_authority_col):
    outliers_summary = {}
    for col in df_numeric.select_dtypes(include=np.number).columns: # we choose only numeric values for finding outliers.
        if col == 'All Fuels':
            continue
        # we will calculate Q1, Q3, and IQR for each column
        Q1 = df_numeric[col].quantile(0.25)
        Q3 = df_numeric[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Determine outliers bound
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers
        outliers_in_col = df_numeric[(df_numeric[col] < lower_bound) | (df_numeric[col] > upper_bound)]
        
        # Get the corresponding local authorities where the outliers exist
        outliers_in_col = outliers_in_col[[local_authority_col, col]]
        
        outliers_summary[col] = {
            'Total Outliers': outliers_in_col.shape[0],  # Count of outliers
            'Outliers Percentage': (outliers_in_col.shape[0] / df_numeric.shape[0]) * 100,  # Percentage of outliers
            'Outliers Data': outliers_in_col  # Local Authority and Outlier Value Data
            }
    
    return outliers_summary

```

```python
local_authority_col = 'Local Authority [Note 2]' 

outliers = detect_outliers_iqr(df_cleaned, local_authority_col)

for column, summary in outliers.items():
    print(f"Outliers in '{column}':")
    print(f"Total Outliers: {summary['Total Outliers']}")
    print(f"Outliers Percentage: {summary['Outliers Percentage']:.2f}%")
    print(f"Outliers Data:\n{summary['Outliers Data']}")
    print("\n")
```

We looked more details about outliers, but I don't remove outliers. Because we have a lot of outliers, and we will lose huge data if we remove outliers. 

Now we will look histogram, the purpose of this visualization is to show the distribution of datasets using histograms. It helps us see how each variable is distributed, understand the overall shape of the data, and spot patterns, outliers, and any potential issues like skewness.

```python
rows = 4  
cols = 4  
plt.figure(figsize=(19, 10))
columns = df_numeric_1.columns
num_cols = len(columns)

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(18, rows * 2), constrained_layout=True)
axes = axes.flatten()  # Flatten to easily iterate through axes

for i, column in enumerate(columns):
    sns.histplot(df_numeric_1[column], kde=True, ax=axes[i], color="blue")
    axes[i].set_title(f" Original Data:{column}")
    axes[i].set_xlabel(column)

plt.show()

```

##### Conclusion

The histograms show that most of the data is concentrated at the lower end, with a few extreme values stretching out towards the higher end. This tells us that the data is heavily skewed, and the outliers seem to be affecting the higher values. Since we're already aware that there are a lot of outliers in our data, this skewness suggests that we need to apply transformations, like a log transformation, to make the data more normal before we dive into further analysis.

### Log Transformation 

```python
rows = 4  
cols = 4  
plt.figure(figsize=(19, 10)) 

# Log transform the data
df_logged = df_numeric_1.apply(lambda x: np.log1p(x))

columns = df_logged.columns
num_cols = len(columns)
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 10), constrained_layout=True)
axes = axes.flatten()

for i, column in enumerate(columns):
    sns.histplot(df_logged[column], kde=True, ax=axes[i], color="Blue")
    axes[i].set_title(f"Log Transformed: \n{column}", fontsize=12, wrap=True) 
    axes[i].set_xlabel('') 

plt.show()

```

##### Conclusion
The log transformation appears to help reduce the skewness in many of the variables. In original distributions, several columns had extreme skew.The log transformation often results in more balanced data. For example, variabla such as "Industrial bioenergy&wastes [Note5]" and "Industrial petroleum", which initially had heavy right tails, now display bell-shaped curves. It shows that our data has been normalized to certain degree. To sum up, the log transformation has successfully decrease the problem of skewness and outliers in the data.

## Additional Map visualization

```python
from geopy.geocoders import Nominatim
import folium
import time

geolocator = Nominatim(user_agent="geoapi", timeout=10) # Geocoding setup with country restriction

# we write function for getting latitude and longitude, restricted to the UK
def geocode_region(region):
    try:
        location = geolocator.geocode(region, country_codes='GB')  # Restrict to UK (GB)
        if location:
            return pd.Series([location.latitude, location.longitude])
    except Exception as e:
        print(f"Error geocoding {region}: {e}")
    return pd.Series([None, None])

df_copy = df_cleaned.copy()

# Apply geocoding to get latitude and longitude
df_copy[['latitude', 'longitude']] = df_copy['Local Authority [Note 2]'].apply(geocode_region)

# Drop rows where geocoding failed
df_copy = df_copy.dropna(subset=['latitude', 'longitude'])

# Get the top 5 places for each region based on "All Fuels"
df_copy_sorted = df_copy.sort_values(by='All Fuels', ascending=False)
df_top_5 = df_copy_sorted.groupby('Local Authority [Note 2]').head(5)

# Create a base map using folium map
map = folium.Map(location=[df_top_5['latitude'].mean(), df_top_5['longitude'].mean()], zoom_start=6)

for _, row in df_top_5.iterrows():
    # Set marker color based condition, if consumtion is greater 400, it is high and marker will be red, other conditions marker will be green. 
    marker_color = "red" if row['All Fuels'] > 400 else "green"
    
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        tooltip=("Click me!"
        ),
        popup=folium.Popup(
            f"<b>Local Authority:</b> {row['Local Authority [Note 2]']}<br>"
            f"<b>Region:</b> {row['Region']}<br>"
            f"<b>All Fuels:</b> {row['All Fuels']:.2f} (ktoe) <br>"
            f"<b>Industrial petroleum:</b> {row['Industrial petroleum']:.2f} (ktoe)",
            max_width=300
        ),
        icon=folium.Icon(color=marker_color),
    ).add_to(map)

map
```

### Conclusion

This map is designed to geographically represent the fuel consumption data in the UK. It highlights the top 5 locations within each region with the highest "All Fuels" consumption, providing a spatial understanding of where fuel usage is most concentrated. This map helps pinpoint hotspots of fuel consumption, with red markers indicating higher consumption levels(>1000). It reveals that some areas have a much higher demand for energy, primarily driven by industrial petroleum. The geospatial context facilitates targeted analysis for energy policy or resource allocation, focusing on regions with the greatest energy needs.

There are following important steps to create map visualization
1. Using the Nominatim geocoder, each "Local Authority" is converted into latitude and longitude coordinates, resdricted to the UK (using country code which is "GB")
2. We filter data in order to get top 5 locations for each region
3. Creating Map, a Folium map is initialized, centered on the mean latitude and longitude of the data
